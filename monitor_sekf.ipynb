```python
import json
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import plotly.tools as tls
import plotly.graph_objects as go
from ipywidgets import interact, widgets, Checkbox, VBox, Output, HBox
from IPython.display import clear_output
```

```python
logfilepattern = "/scratch/fac2/hm_home/carra2_@stream@/monitoring/SEKF.json"
streams = ['198409', '198909', '199409', '199909', '200409', '200909', '201409', '201909']


const = {
    "ANA_INCR1": {
        "variable": "WG6",
        "bins":(-.1,.1)},
    "ANA_INCR2": {
        "variable": "WG5",
        "bins":(-.1,.1)},
    "ANA_INCR3": {
        "variable": "WG4",
        "bins": (-.1,.1)},
    "ANA_INCR4": {
        "variable": "WG3",
        "bins":(-.1,.1)},
    "ANA_INCR5": {
        "variable": "WG2",
        "bins":(-.1,.1)},
    "ANA_INCR6": {
        "variable": "TG2",
        "bins":(-1,1)},
    "ANA_INCR7": {
        "variable": "TG1",
        "bins":(-1,1)}
}

varnames = list(const.keys())

data_dict = {}
for i in range(len(streams)):
    logfile = logfilepattern.replace("@stream@", streams[i])
    with open(logfile) as f:
       data_dict[streams[i]] = json.load(f)
```

```python

out = Output()

def plot_graph(variable_name):
    # Clear the previous plot
    #print(stream, variable_name)
    with out:
        clear_output(wait=True)
        fig = go.Figure()
        for stream, checkbox in stream_checkboxes.items():
            if checkbox.value:
                data0 = data_dict[stream][variable_name]
                for metric, checkbox in metric_checkboxes.items():
                    if checkbox.value:
                        fig.add_trace(go.Scatter(x=data0['analysis_time'], y=data0[metric], name=stream + '_' + metric))
        fig.show()

# Assuming that all variables in your dictionary have the same metrics
sample_variable = next(iter(data_dict.values()))  # X001...
variables = sample_variable.keys()
metrics = next(iter(sample_variable.values()))

stream_selector = widgets.Dropdown(
    options=streams,
    description='Stream:',
)

variable_selector = widgets.Dropdown(
    options=variables,
    description='Variable:',
)

metric_checkboxes = {metric: Checkbox(value=False, description=metric) for metric in metrics}
stream_checkboxes = {stream: Checkbox(value=False, description=stream) for stream in streams}

def update_plot(change):
    plot_graph(variable_selector.value)

variable_selector.observe(update_plot, names='value')
#stream_selector.observe(update_plot, names='value')

for checkbox in metric_checkboxes.values():
    checkbox.observe(update_plot, names='value')

for checkbox in stream_checkboxes.values():
    checkbox.observe(update_plot, names='value')


display(HBox(list(stream_checkboxes.values())), variable_selector, HBox(list(metric_checkboxes.values())) , out)


```
